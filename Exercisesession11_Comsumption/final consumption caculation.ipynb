```python
import csv
```

```python
class Unit:
    def __init__(self, unit,x,y,food,frozen_food,office_supplies,furniture,daily_necessities):
        self.unit=unit
        self.x=x
        self.y=y
        self.food=food
        self.frozen_food=frozen_food
        self.office_supplies=office_supplies
        self.furniture=furniture
        self.daily_necessities=daily_necessities
```

```python
units = []
with open("final result.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        unit = row['unit']
        x=row['x']
        y=row['y']
        food=row['food']
        frozen_food = row['frozen food']
        office_supplies=row['office supplies']
        furniture=row['furniture']
        daily_necessities=row['daily necessities']
        units.append(Unit(unit,x,y,food,frozen_food,office_supplies,furniture,daily_necessities))
```

```python
f=[]
ff=[]
o=[]
fn=[]
d=[]
n=10
locationX=[]
locationY=[]
minf=0
minff=0
mino=0
minfn=0
mind=0

for u in units:
    f.append(float(u.food))
    ff.append(float(u.frozen_food))
    o.append(float(u.office_supplies))
    fn.append(float(u.furniture))
    d.append(float(u.daily_necessities))
    locationX.append(int(u.x))
    locationY.append(int(u.y))

for i in sorted(f):
    if i>0:
        minf=i
        break

for i in sorted(ff):
    if i>0:
        minff=i
        break

for i in sorted(o):
    if i>0:
        mino=i
        break

for i in sorted(fn):
    if i>0:
        minfn=i
        break

for i in sorted(d):
    if i>0:
        mind=i
        break


sf=(max(f)-minf)/n
sff=(max(ff)-minff)/n
so=(max(o)-mino)/n
sfn=(max(fn)-minfn)/n
sd=(max(d)-mind)/n



print(sf)
print(sff)
```
