```python
import numpy as np
from numpy import unique
from numpy import where
from sklearn.datasets import make_classification
from sklearn.cluster import KMeans
from matplotlib import pyplot
import pandas as pd
import csv
```

```python
df = pd.read_csv( 'WorldCities.csv')
x = df
x_np = np.array(x)  
```

```python
n_clusters = 5                 
estimator = KMeans(n_clusters)
```

```python
estimator.fit(x)
```

```python
markers = ['*', 'v', '+', '^', 's']      # List of markup styles
colors = ['r', 'g', 'm', 'c', 'orange']  # List of marker colors
labels = estimator.labels_      # Get cluster labels

pyplot.figure(figsize=(9, 6))
pyplot.title("Clustering of cities in the world ", fontsize=25)
pyplot.xlabel('Longitude', fontsize=18)
pyplot.ylabel('Latitude', fontsize=18)

for i in range(n_clusters):    # Iterate over all coordinate points and draw scatter plots
    members = labels == i      # Members is a Boolean array  
    pyplot.scatter(
        x_np[members, 1],      # An array of longitude
        x_np[members, 0],      # An array of latitude
        marker = markers[i],   # Mark style
        c = colors[i]          # Mark color
    )   

pyplot.grid()
pyplot.show()
```
